```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv(r'C:\Users\ASUS\Downloads\current-power.csv')

df
```

```python
df.cov()
```

```python
df.describe()
```

```python
df.corr()
```

```python
x = df['Current']
y = df['Power']

plt.figure()
plt.scatter(x,y)
plt.title('Device Currentt vs Power Response')
plt.xlabel('Power (w)')
plt.ylabel('Current (A)')
plt.show()
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope * x + intercept

plt.figure(figsize=(10,5))
plt.scatter(
    x,y,
    color = '#1c1f24',
    alpha=0.6,
    label = 'Data points'
)

plt.plot(
    x,
    line,
    color = 'blue',
    label =f'Regression line (r = {r_value:.3f})'
)

plt.title('Device Currentt vs Power Response')
plt.xlabel('Power (w)')
plt.ylabel('Current (A)')
plt.legend()
plt.show()
```

## Air Quality <hr style = "border:2.5px solid #ff0000ff"></hr>

```python
air = pd.read_csv(r'C:\Users\ASUS\Downloads\air-quality.csv', delimiter = ",")

air
```

## NOX <hr style = "border:2.5px solid #ff0000ff"></hr>

```python
x = air['NOx']
y = air['Relative_Humidity']

slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope * x + intercept

plt.figure(figsize=(10,5))
plt.scatter(
    x,y,
    color = '#1c1f24',
    alpha=0.2,
    label = 'Data points'
)

plt.plot(
    x,
    line,
    color = 'blue',
    label =f'Regression line (r = {r_value:.3f})'
)

plt.title('NOx vs Humidity Correlation')
plt.xlabel('Relative Humidity (%)')
plt.ylabel('NOx (ug/cu.m)')
plt.legend()
plt.show()
```
